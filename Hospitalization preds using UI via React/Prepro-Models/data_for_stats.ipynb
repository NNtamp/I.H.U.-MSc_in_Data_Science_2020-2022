```python
import pandas as pd
data = pd.read_csv('data.csv')
data=data.set_index('patient_id')
```

```python
import numpy as np
```

```python
list_str=[]
column_list = data.columns.values.tolist()
for column_name in column_list:
    try:
        if not data.loc[0,column_name].isdigit():
            list_str.append(column_name)
    except: continue

for j in list_str:
    for i in range(data.shape[0]):
        try:data.loc[i,j]=data.loc[i,j].strip()
        except:continue
```

```python
def insurance(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'insurance'] != data.loc[i,'insurance']) or (""==data.loc[i,'insurance']) or ("?"==data.loc[i,'insurance']):
                data.loc[i,'insurance']='NAN'        
        except:continue
    return (data)     
    
```

```python
def num_diagnoses(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'num_diagnoses'] <=139) :
                data.loc[i,'num_diagnoses']='GROUP_001_139'
            elif (data.loc[i,'num_diagnoses'] <=239) :
                data.loc[i,'num_diagnoses']='GROUP_140_239'
            elif (data.loc[i,'num_diagnoses'] <=279) :
                data.loc[i,'num_diagnoses']='GROUP_240_279' 
            elif (data.loc[i,'num_diagnoses'] <=289) :
                data.loc[i,'num_diagnoses']='GROUP_280_289'
            elif (data.loc[i,'num_diagnoses'] <=319) :
                data.loc[i,'num_diagnoses']='GROUP_290_319'
            elif (data.loc[i,'num_diagnoses'] <=389) :
                data.loc[i,'num_diagnoses']='GROUP_320_389'
        except:continue
    return (data)     
    
```

```python
num_diagnoses(data)
```

```python
data['num_diagnoses'].unique()
```

```python
def hospitalization(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'hospitalization'] <1.00) :
                data.loc[i,'hospitalization']='DAY'
            elif (data.loc[i,'hospitalization'] >=1.00) and  (data.loc[i,'hospitalization'] <7.00) :
                data.loc[i,'hospitalization']='WEEK'
            elif (data.loc[i,'hospitalization'] >=7.00) and  (data.loc[i,'hospitalization'] <14.0) :
                data.loc[i,'hospitalization']='TWO_WEEKS' 
            elif (data.loc[i,'hospitalization'] >=14.00) and  (data.loc[i,'hospitalization'] <30.0) :
                data.loc[i,'hospitalization']='MONTH'
            elif (data.loc[i,'hospitalization'] >=30.0) :
                data.loc[i,'hospitalization']='MORE'
        except:continue
    return (data)  
```

```python
hospitalization(data)
```

```python
data['hospitalization'].unique()
```

```python
def age(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'age'] <=0) :
                data.loc[i,'age']='NEWBORN'
            elif (data.loc[i,'age'] <=20) :
                data.loc[i,'age']='YOUNG' 
            elif (data.loc[i,'age'] <=40) :
                data.loc[i,'age']='ADULTS'
            elif (data.loc[i,'age'] <=60) :
                data.loc[i,'age']='MIDDLE AGED'
            elif (data.loc[i,'age'] >60) :
                data.loc[i,'age']='ELDERS'
                
        except:continue
    return (data)  
```

```python
def religion(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'religion'] =='NOT SPECIFIED') or (data.loc[i,'religion'] =='UNOBTAINABLE') or (data.loc[i,'religion']=='OTHER'):
                data.loc[i,'religion']='NOT SPECIFIED'
            elif (data.loc[i,'religion'] =='PROTESTANT QUAKER') or (data.loc[i,'religion'] =='JEWISH')  or (data.loc[i,'religion'] =='CATHOLIC') or (data.loc[i,'religion'] =='EPISCOPALIAN') or (data.loc[i,'religion'] =='SCIENTOLOGY') or (data.loc[i,'religion'] =='CHRISTIAN SCIENTIST') or (data.loc[i,'religion'] =='BUDDHIST') or (data.loc[i,'religion'] =="JEHOVAH'S WITNESS") or (data.loc[i,'religion'] =="GREEK ORTHODOX") or (data.loc[i,'religion'] =='MUSLIM') or (data.loc[i,'religion'] =='STAN') or (data.loc[i,'religion'] =='7TH DAY ADVENTIST') or (data.loc[i,'religion'] =='UNITARIAN-UNIVERSALIST')  or (data.loc[i,'religion'] =='ROMANIAN EAST. ORTH') or (data.loc[i,'religion'] =='BAPTIST')  or (data.loc[i,'religion'] =='HINDU')  or (data.loc[i,'religion'] =='HEBREW') or (data.loc[i,'religion'] =='METHODIST'):
                data.loc[i,'religion']='RELIGIOUS' 
            elif (data.loc[i,'religion'] != data.loc[i,'religion']) :
                     data.loc[i,'religion']='NAN'
        except:continue
    return (data) 
```

```python
def marital_status(data):
    for i in range(data.shape[0]):
        try:
            if (data.loc[i,'marital_status'] =='DIVORCED') or (data.loc[i,'marital_status'] =='WIDOWED') or (data.loc[i,'marital_status'] =='SEPARATED'):
                data.loc[i,'marital_status']='DIVORCED/WIDOWED/SEPARATED'
            elif (data.loc[i,'marital_status'] =='MARRIED') or (data.loc[i,'marital_status'] =='LIFE PARTNER') :
                data.loc[i,'marital_status']='LIFE PARTNER/MARRIED'
            elif (data.loc[i,'marital_status'] != data.loc[i,'marital_status']) or ("?"==data.loc[i,'marital_status']):
                data.loc[i,'marital_status']='NAN'
            elif (data.loc[i,'marital_status'] == 'UNKNOWN (DEFAULT)') :
                data.loc[i,'marital_status']='OTHER'
        except:continue
    return (data) 
```

```python
def ethnicity(data):
    for i in range(data.shape[0]):
        try:
            if ("WHITE" in data.loc[i,'ethnicity']) or ("CAUCA" in data.loc[i,'ethnicity']) or ("UNK"==data.loc[i,'ethnicity']) or ("SOUTH AMERICAN"==data.loc[i,'ethnicity']) or ("PORTUGUESE"==data.loc[i,'ethnicity']):
                data.loc[i,'ethnicity']='WHITE'
            elif ("BLACK" in data.loc[i,'ethnicity']):
                data.loc[i,'ethnicity']='BLACK'
            elif ("GUERO"==data.loc[i,'ethnicity']) or ("HISPANIC" in data.loc[i,'ethnicity']):
                data.loc[i,'ethnicity']='HISPANIC/LATINO'            
            elif (data.loc[i,'ethnicity'] != data.loc[i,'ethnicity']) or ("?" in data.loc[i,'ethnicity']) or (""==data.loc[i,'ethnicity']) or(np.nan==data.loc[i,'ethnicity']) or (pd.isna(data.loc[i,'ethnicity'])):
                data.loc[i,'ethnicity']='NAN'
            elif ('ASIAN' in data.loc[i,'ethnicity']):
                data.loc[i,'ethnicity']='ASIAN'
            else:
                data.loc[i,'ethnicity']='OTHER'
        except: continue
            
    data['ethnicity']=pd.Series(data['ethnicity'], dtype=object).fillna('NAN').tolist()
    return(data)
```
